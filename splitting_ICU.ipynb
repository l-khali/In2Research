```python
import numpy as np
import pandas as pd
from visits import Visit, Cohort, MissingDataError, holm_bonferroni, icu_split
import glob
import pymannkendall as mk
from datetime import datetime, timedelta
from scipy.stats import mannwhitneyu, ttest_ind
```

```python
cohort_1_details = pd.read_csv("cohort_1_details_v5.csv", sep=',')
cohort_2_details = pd.read_csv("cohort_2_details_v5.csv", sep=',')
cohort_3_details = pd.read_csv("cohort_3_details_v5.csv", sep=',')
cohort_4_details = pd.read_csv("cohort_4_details_v5.csv", sep=',')
```

# Variance HR

```python
variance_hr_results_c1 = pd.read_csv("variance_results_c1.csv", sep=",")
variance_hr_results_c2 = pd.read_csv("variance_results_c2_entire.csv", sep=",")
```

```python
proportions_c1, results_c1, failed_c1 = icu_split(variance_hr_results_c1, cohort_1_details, cohort_2_details, cohort_3_details, cohort_4_details);
```

```python
proportions_c2, results_c2, failed_c2 = icu_split(variance_hr_results_c2, cohort_1_details, cohort_2_details, cohort_3_details, cohort_4_details)
```

```python
for df_c1, df_c2 in zip(results_c1, results_c2):
    print(mannwhitneyu(df_c1["tau"], df_c2["tau"], alternative="greater"))
    print(ttest_ind(df_c1["tau"], df_c2["tau"], equal_var=False))
```
